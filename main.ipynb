python Lists

```python
vegetables = ["cucmber", "tomato","potato"]
print(vegetables[1])
print(vegetables[0:1])
print(vegetables)
#Добавление в Lists
vegetables.append("orange")
print(vegetables)
#Добавление в Lists, но на конкретное место
vegetables.insert(2, "ananas")
print(vegetables)
#Удаление чего-то из спика
vegetables.pop(1)
print(vegetables)
#Удавление всего содержимого списка
#vegetables.clear()
#print(vegetables)
```

Переборка в цикле

```python
for i in range(len(vegetables)):
    print(vegetables[i])
vegetables.sort()
print(vegetables)
```

Классы и обращение к ним


```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p1 = Person("Peter", "25")
print(p1.name + " " + p1.age) 

```
